# **Tiền xử lý dữ liệu**

```python
# Bỏ các cột không quan trọng
df = car.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style',
               'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
# Đổi tên cột
df = df.rename(columns={'Engine HP': 'Mã lực', 'Engine Cylinders': 'Xi-lanh',
                         'Transmission Type': 'Truyền tải', 'Driven_Wheels': 'Mode',
                         'highway MPG': 'Tiêu thụ ko ngừng',
                         'city mpg': 'Tiêu thụ thường', 'MSRP': 'Giá thành'})
df.head()
```

```python
print(df.shape,'\n') # Số hàng và cột
dup_car = df[df.duplicated()] # Tìm các thông tin trùng lắp
print(dup_car.shape,'\n')

print(df.count(),'\n') # Đếm các hàng không rỗng từng cột
df = df.drop_duplicates() # Xóa các thông tin trùng lắp
print(df.count(), '\n')

print(df.isnull().sum()) # Đếm các thông tin rỗng từng cột
df = df.dropna() # Xóa các hàng rỗng
print(df.count(), '\n')
```

# **Phân tích dữ liệu**

```python
car = df
sns.boxplot(x=car['Giá thành'])
```

```python
sns.boxplot(x=car['Mã lực'])
```

```python
# Biểu đồ số xe mỗi thương hiệu
car.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
# Đặt tiêu đề
plt.title('Số lượng sản xuất')
plt.xlabel('Thương hiệu')
plt.ylabel('Số lượng xe')
```

# **Phân tích mối tương quan**

```python
plt.figure(figsize=(20,10))
c = car.corr() # Mối liên hệ
sns.heatmap(c, cmap='BrBG', annot=True) # Biểu đồ thể hiện độ tương quan
```

```python
df = df.rename(columns={'Mã lực':'HP', 'Xi-lanh': 'Cylinder'})
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (15, 7)
plt.title('Mã lực - Xilanh')

sns.scatterplot(x=df.Cylinder, y=df.HP, hue=df.Make, s=200)
plt.legend(loc='upper left', fontsize='10')
plt.ylabel('Mã lực')
plt.xlabel('Xi-lanh')
```

```python
df_hp = df.groupby('Make')['HP'].sum()
df_hp.plot.pie(autopct='%1.1f%%')
plt.title('Mã lực - Thương hiệu')
plt.ylabel('')
```

```python
# Tạo một bảng ánh xạ giữa hãng xe và quốc gia
make_country_mapping = {
    'BMW': 'Germany',
    'Audi': 'Germany',
    'FIAT': 'Italy',
    'Mercedes-Benz': 'Germany',
    'Chrysler': 'USA',
    'Nissan': 'Japan',
    'Volvo': 'Sweden',
    'Mazda': 'Japan',
    'Mitsubishi': 'Japan',
    'Ferrari': 'Italy',
    'Alfa Romeo': 'Italy',
    'Toyota': 'Japan',
    'McLaren': 'UK',
    'Maybach': 'Germany',
    'Pontiac': 'USA',
    'Porsche': 'Germany',
    'Saab': 'Sweden',
    'GMC': 'USA',
    'Hyundai': 'South Korea',
    'Plymouth': 'USA',
    'Honda': 'Japan',
    'Oldsmobile': 'USA',
    'Suzuki': 'Japan',
    'Ford': 'USA',
    'Cadillac': 'USA',
    'Kia': 'South Korea',
    'Bentley': 'UK',
    'Chevrolet': 'USA',
    'Dodge': 'USA',
    'Lamborghini': 'Italy',
    'Lincoln': 'USA',
    'Subaru': 'Japan',
    'Volkswagen': 'Germany',
    'Spyker': 'Netherlands',
    'Buick': 'USA',
    'Acura': 'Japan',
    'Rolls-Royce': 'UK',
    'Maserati': 'Italy',
    'Lexus': 'Japan',
    'Aston Martin': 'UK',
    'Land Rover': 'UK',
    'Lotus': 'UK',
    'Infiniti': 'Japan',
    'Scion': 'Japan',
    'Genesis': 'South Korea',
    'HUMMER': 'USA',
    'Bugatti': 'France'
}

# Thêm cột 'Country' dựa vào cột 'Make'
df['Country'] = df['Make'].map(make_country_mapping)
df
```

```python
df_hp = df.groupby('Country')['HP'].sum()
df_hp.plot.pie(autopct='%1.1f%%')
plt.title('Mã lực - Quốc gia')
plt.ylabel('')
```

```python
plt.rcParams['figure.figsize'] = (15, 7)
plt.title('Mã lực - Xilanh')
sns.scatterplot(x=df.Cylinder, y=df.HP, hue=df.Country, s=200)
plt.legend(loc='upper left', fontsize='10')
plt.ylabel('Mã lực')
plt.xlabel('Xi-lanh')
```

```python
cor = df.corr(method='pearson')
mask = np.triu(np.ones_like(cor, dtype=bool))

f, ax = plt.subplots(figsize=(10, 5))
sns.heatmap(cor, mask=mask, cmap='Blues', square=True, ax=ax)
```

```python
#df.columns.tolist()
df_gt = df.rename(columns={'Giá thành': 'Price'})
df_gt = df_gt.groupby('Country')[['Price']].mean()

plt.rcParams['figure.figsize'] = (12, 8)
plt.title('Quốc gia - Giá thành')
plt.ylabel('Giá thành', fontsize=15)
plt.xlabel('Quốc gia', fontsize=15)
plt.xticks(rotation=30, ha='right')
plt.bar(df_gt.index, df_gt.Price)
```
